```python
### Rozkład prawdopodobieństwa – miara probabilistyczna określona na zbiorze wartości pewnej zmiennej losowej 
### (wektora losowego), przypisująca prawdopodobieństwa wartościom tej zmiennej.

### Funkcja gęstości prawdopodobieństwa (probability density function) - odpowiada prawdopodobieństwu wystąpienia wartości 
### z danego zakresu wartości.

### Funkcja masy prawdopodobieństwa (prabobility mass function) - funkcja dająca dla każdej liczby rzeczywistej 
### prawdopodobieństwo, że dana dyskretna zmienna losowa przyjmie tą wartość.
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, binom, poisson
```

```python
### Rozkład jednostajny ciągły (uniform data distribution)- ciągły rozkład prawdopodobieństwa, dla którego gęstość 
### prawdopodobieństwa w przedziale od a do b jest stała i różna od zera, a poza nim równa zeru.

values = np.random.uniform(-10, 10, 10000)
plt.hist(values, 50)
plt.show()
```

```python
### Rozkład normalny, rozkład Gaussa (normal distribution) - jeden z najważniejszych rozkładów prawdopodobieństwa, 
### odgrywający ważną rolę w statystyce. Wykres funkcji prawdopodobieństwa tego rozkładu jest krzywą w kształcie dzwonu 
## (tak zwaną krzywą dzwonową).

'''
Przyczyną jego znaczenia jest częstość występowania w naturze. Jeśli jakaś wielkość jest sumą lub średnią bardzo wielu 
drobnych losowych czynników, to niezależnie od rozkładu każdego z tych czynników jej rozkład będzie zbliżony do normalnego
(centralne twierdzenie graniczne) – dlatego można go bardzo często zaobserwować w danych.
'''

xValues = np.arange(-3, 3, 0.001)
plt.plot(xValues, norm.pdf(xValues))
```

```python
### Rozkład wykładniczy (Exponential distribution function) - gdy jesteśmy blisko zera to prawdopodobieństwo jest duże ale odlalając
### się zmniejsza się wykładniczo

xValues = np.arange(0, 10, 0.001)
plt.plot(xValues, expon.pdf(xValues))
```

```python
### Rozkład dwumianowy (Binomial distribution) – dyskretny rozkład prawdopodobieństwa opisujący liczbę sukcesów k w ciągu 
### N niezależnych prób, z których każda ma stałe prawdopodobieństwo sukcesu równe p.

n, p = 10, 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x,n,p))
```

```python
### Rozkład Poissona (Poisson distribution) - dyskretny rozkład prawdopodobieństwa, wyrażający prawdopodobieństwo szeregu 
### wydarzeń mających miejsce w określonym czasie, gdy te wydarzenia występują ze znaną średnią częstotliwością i w sposób 
### niezależny od czasu jaki upłynął od ostatniego zajścia takiego zdarzenia. Rozkład Poissona można również stosować w 
### odniesieniu do liczby zdarzeń w innych określonych przedziałach, takich jak odległość, powierzchnia lub objętość.

### Funkcja masy prawdopodobieństwa

mu = 80
x = np.arange(50, 100, 0.5)
plt.plot(x, poisson.pmf(x, mu))
```

```python
### Kowariancja (covariance) - liczba określająca zależność liniową między zmiennymi losowymi X i Y
```

```python
### Korelacja (correlation) - związek pomiędzy dwiema zmiennymi losowymi X i Y.
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

sales = np.random.normal(0, 20, 1000)
timeSpent = np.random.normal(0, 10, 1000)

plt.scatter(sales, timeSpent)
plt.show()

np.corrcoef(sales, timeSpent)
```

```python
sales = (timeSpent * 10) - 10

plt.scatter(sales,timeSpent)
plt.show()

np.corrcoef(sales, timeSpent)
```
